```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\aksha\Downloads\Universities.csv",index_col="Univ")
```

```python
df
```

```python
std=StandardScaler()
```

```python
df1=pd.DataFrame(std.fit_transform(df),columns=df.columns)
```

```python
df1
```

```python
df1.info()
```

### create the model

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=6)#here we mention 6 then we will get 6 pca(6 lines)
```

```python
model.fit_transform(df1)
```

```python
var=model.explained_variance_ratio_
var#here the 6 numbers are 6 pca and its varience
```

```python
np.cumsum(var)
```

```python
#here we select frist 3 pca with total varienc=94.7 %
```

```python
final_model=PCA(n_components=3)
final_model.fit_transform(df1)
```
