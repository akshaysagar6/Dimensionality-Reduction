## HIERARCHY CLUSTERING

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as sch
```

```python
dendrogram=sch.dendrogram(sch.linkage(new_df,'complete'))
```

```python
dendregram=sch.dendrogram(sch.linkage(new_df,'average'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(new_df,'centroid'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(new_df,'ward'))
```

```python
n_clst=[2,3,4,5,6,7,8,9,10]
link=['ward','complete','average','single']
output=[]

for i in n_clst:
    for j in link:
        agg=AgglomerativeClustering(n_clusters=i,linkage=j)
        clusters=agg.fit(pca_df)
        cl=clusters.labels_
        silscore=silhouette_score(pca_df,cl)
        print('cluster',i,'linkage',link,'silhouette score',silscore)
        output.append([i,j,silscore])
        
```

```python
output
```

```python
h_df=pd.DataFrame(output,columns=('clusters','linkage','score'))
h_df
```

```python
h_df['score'].max()
```

```python
h_df[h_df.score==0.4465449278023583]
```

## KMEANS CLUSTERING

```python
from sklearn.cluster import KMeans
```

## optimum  value for kmeans

```python
output=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(new_df)
    output.append(kmeans.inertia_)
```

```python
output
```

```python
plt.plot(range(1,10),output)
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.title('optimum value')
```
