```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\aksha\Downloads\iris.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
features=df.drop('variety',axis=1)
```

```python
features
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
target=df[['variety']]
```

```python
target=label.fit_transform(target)
```

```python
target
```

### Create The Model

```python
from sklearn.manifold import TSNE
```

```python
model=TSNE(n_components=2)
```

```python
tsne_comp=model.fit_transform(features)
tsne_comp
```

```python

plt.scatter(tsne_comp[:,0],tsne_comp[:,1],c=target,edgecolors='b')
plt.show()
```
