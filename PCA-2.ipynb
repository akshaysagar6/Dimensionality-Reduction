```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
std=StandardScaler()
```

```python
df1=pd.DataFrame(std.fit_transform(df),columns=df.columns)
```

```python
df1
```

```python
features=df1.drop('target',axis=1)
```

```python
features
```

```python
model=PCA(n_components=30)
```

```python
model.fit_transform(features)
```

```python
var=model.explained_variance_ratio_
var
```

```python
np.cumsum(var)
```

```python
final_model=PCA(n_components=7)
```

```python
final_model.fit_transform(features)
```

### here we reduced the number of features to 7 from 30
